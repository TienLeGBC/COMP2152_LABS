```csharp
"""
Author: Le Tram Nhat Tien
Assignment: #1
"""

#B
gym_member = "Tien Le"
preferred_weight_kg = 85.5
highest_reps = 35 
active_member = True

#C (Dictionary with keys as friend names (str) and values as tuples of workout minutes (int, int, int))
workout_stats = {
    "Kim Chi": (30, 45, 20),
    "Phuc": (25, 35, 40),
    "Vu": (40, 30, 15)
}

#D
for friend, minutes in list(workout_stats.items()):  # chuyá»ƒn items() sang list()
    total = sum(minutes)
    workout_stats[f"{friend}_Total"] = total
print(workout_stats)


#E
workout_list = [list(minutes) for name, minutes in workout_stats.items() if not name.endswith("_Total")]

print(workout_list)

#F
yoga_running = [row[:2] for row in workout_list]
print("Yoga + Running minutes:", yoga_running)


weightlifting_last_two = [row[2] for row in workout_list[-2:]]
print("Weightlifting minutes (last 2 friends):", weightlifting_last_two)


#G
for friend, total in workout_stats.items():
    if friend.endswith("_Total") and total >= 120:
        name = friend.replace("_Total", "")
        print(f"Great job staying active, {name}!")


#H
friend_name = input("Enter friend's name: ")

if friend_name in workout_stats:
    minutes = workout_stats[friend_name]
    total = workout_stats[f"{friend_name}_Total"]
    print(f"{friend_name} workout minutes: Yoga={minutes[0]}, Running={minutes[1]}, Weightlifting={minutes[2]}, Total={total}")
else:
    print(f"Friend {friend_name} not found in the records.")


#I
totals = {k.replace("_Total",""): v for k,v in workout_stats.items() if k.endswith("_Total")}
max_friend = max(totals, key=totals.get)
min_friend = min(totals, key=totals.get)

print(f"Friend with highest total workout minutes: {max_friend} ({totals[max_friend]} minutes)")
print(f"Friend with lowest total workout minutes: {min_friend} ({totals[min_friend]} minutes)")




```
